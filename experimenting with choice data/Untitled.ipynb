```python
#!wget "http://individual.utoronto.ca/vaguirre/wpapers/aguirregabiria_mira.zip"
#!unzip "aguirregabiria_mira.zip" -d 'aguirregabiria_mira'
```

```python
#https://editorialexpress.com/jrust/crest_lectures/zurcher.pdf

#https://www.kaggle.com/datasets/erichschulman/bus1234.csv?resource=download
```

```python
def addmonthyear(x):
    x =x.copy()
    x['monthyear'] =(x.year - x.year.min())*12 +x.month

    x['time_index'] = x['monthyear'] - x['monthyear'].min()

    return x
    
df = pd.read_csv('aguirregabiria_mira/bus1234.csv')
df = df.sort_values(['id','year','month'])#.groupby(['id']).apply( lambda x: (x.year - x.year.min())*12 +x.month   )


df = df.groupby(['id']).apply( lambda x: addmonthyear(x) ).reset_index(drop=True).drop(columns = ['Unnamed: 0'])

group_df = pd.get_dummies(df['group']).astype(int)
group_df.columns = [ 'group_'+str(i)  for i in  range(df['group'].nunique())]

group_df = group_df.drop(columns = 'group_3')

df = pd.concat( [ df, group_df],axis = 1)

df['obs'] = df[['year','month','miles','time_index','group_0','group_1','group_2'#,'group_3'
                                                       ]].apply(lambda x : list(x.values),axis =1 )
df['actions'] = df['replace'].astype(int)
df['rewards'] = 0 

df['episode_starts']  = (df.id != df.id.shift(1)).astype(int)


df.to_csv('final/aguirregabiria_mira.csv')
df.to_pickle('final/aguirregabiria_mira.pkl')
```
